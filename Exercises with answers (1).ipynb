#Exercise 4


Write a function that takes a prediction (can be 1 or 0) and a label (1 or 0) as parameters. The function should return:


*   "tp" if the prediction is a true positive
*   "tn" if the prediction is a true negative
*   "fn" if the prediction is a false negative
*   "fp" if the prediction is a false positive

Test your function with various input

````
def single_event_classification(pred, label):
    if pred==1 and label==1:
        return "tp"
    if pred==1 and label==0:
        return "fp"
    if pred==0 and label==1:
        return "fn"
    if pred==0 and label==0:
        return "tn"
````

````
print(single_event_classification(0, 0))
print(single_event_classification(0, 1))
print(single_event_classification(1, 0))
print(single_event_classification(1, 1))
````

````
from random import random
true_labels = [1 if random() > 0.5 else 0 for x in range(0, 201)]
predictions = [x if random() > 0.15 else abs(1-x) for x in true_labels]
````

Write a function that takes the lists of predictions and labels and creates and returns a dictionary that contains the counts of true positives, true negatives, false positives, false negatives for the input lists. Your function should call the function you wrote for the last task.  

Test your function using `true_labels` and `predictions` variables.

````
def count_event_types(preds, labels):
    event_counts = {}
    for i in range(len(preds)):
        event_type = single_event_classification(preds[i], labels[i])
        if event_type not in event_counts:
            event_counts[event_type] = 0
        event_counts[event_type] += 1
    return event_counts
````

Write a function that takes the lists of predictoins and labels and returns the accuracy of the predictions. Your function should call the function that counts the types of events (from the last exercise).

Test your function using `true_labels` and `predictions` variables.

````
def compute_accuracy(preds, labels):
    event_dict = count_event_types(preds, labels)
    return (event_dict["tp"] + event_dict["tn"]) / (event_dict["tp"] + event_dict["tn"] + event_dict["fp"] + event_dict["fn"])
    
````

Optional:

Write a function that computes the MCC metric (refer for the formula: https://en.wikipedia.org/wiki/Matthews_correlation_coefficient).
You will need to use the `math` module to calculate the square root for the denominator.

Test your function using `true_labels` and `predictions` variables.

````
import math
def compute_mcc(preds, labels):
    event_dict = count_event_types(preds, labels)
    num = (event_dict["tp"] * event_dict["tn"]) - (event_dict["fp"] * event_dict["fn"])
    denom = (event_dict["tp"] + event_dict["fp"]) * (event_dict["tp"] + event_dict["fn"])
    denom *= (event_dict["tn"] + event_dict["fp"]) * (event_dict["tn"] + event_dict["fn"])
    denom = math.sqrt(denom)
    return num / denom
````

````
compute_mcc(predictions, true_labels)
````

Code to load file:

````
# The following is code for uploading a file to the colab.research.google 
# environment.

# library for uploading files
from google.colab import files 

def upload_files():
    # initiates the upload - follow the dialogues that appear
    uploaded = files.upload()

    # verify the upload
    for fn in uploaded.keys():
        print('User uploaded file "{name}" with length {length} bytes'.format(
            name=fn, length=len(uploaded[fn])))

    # uploaded files need to be written to file to interact with them
    # as part of a file system
    for filename in uploaded.keys():
        with open(filename, 'wb') as f:
            f.write(uploaded[filename])
upload_files()
````

Write your code here:

````
experimental_results = {}

with open("experiments.txt", "r") as input_file:
  for line in input_file:
    if line.startswith("#"):
      experiment_name = line[14:].strip()
      experimental_results[experiment_name] = ([], [])
    else:
      _, pred, label = line.split('\t')
      experimental_results[experiment_name][0].append(int(pred))
      experimental_results[experiment_name][1].append(int(label))

results_list = []
for experiment in experimental_results:
  preds, labels = experimental_results[experiment]
  results_list.append((experiment, compute_mcc(preds, labels)))

results_list.sort(key=lambda tuple: tuple[1], reverse=True)

for experiment, mcc in results_list:
  print(f"Experiemnt: {experiment}, results are: {mcc}")
````
