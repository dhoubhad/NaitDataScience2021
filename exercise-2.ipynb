Use a form to give the user three options. The options should be "short", "medium", and "long". When the user selects "short" print "Hello world!". When the user selects "medium" print "How are you doing, world?!". When the user selects "long" print "Hello world, we all love data science!!". Use if/elif statements to accomplish this.

Bonus: Your program is set up to handle three options. What should you do if an unexpected option is encountered?

```
options = "Long" #@param ["Small", "Medium", "Long"]

if options == "Small":
  print("Hello world!")
elif options == "Medium":
  print("How are you doing, world?")
elif options == "Long":
  print("Hello world, we all love data science!!")
```

Your program should print two different counts. Considering the numbers 1 to 100, your program should count the number of odd numbers and the number of numbers divisible by 7. Your program must:


*   Use the range function
*   Use a for loop



```
count_odd = 0
count_div_7 = 0
for each in range(1,100):

  if each%2 == 1:
    count_odd += 1
  if each%7 == 0:
    count_div_7 += 1

print('Count of odd numbers:', count_odd)
print('Count of numbers divisible by 7:', count_div_7)
```
