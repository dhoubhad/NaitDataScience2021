Write a function that takes a prediction (can be 1 or 0) and a label (1 or 0) as parameters. The function should return:


*   "tp" if the prediction is a true positive (prediction and label are both 1)
*   "tn" if the prediction is a true negative (prediction and label are both 0)
*   "fn" if the prediction is a false negative (prediction is 0 but label is 1)
*   "fp" if the prediction is a false positive (prediction is 1 but label is 0)


```
def test_func(prediction, label):
  if (prediction and label) == 1:
    return "tp"
  elif (prediction or label) == 0:
    return "tn"
  elif prediction == 0 and label == 1:
    return "fn"
  elif prediction == 1 and label == 0:
    return "fp"
```

Test your function with various input

```
test_func(1,0)
```

Below we provided you with the code to generate some data to work with. Please run the cell below and move to the next one

```
from random import random
true_labels = [1 if random() > 0.5 else 0 for x in range(0, 201)]
predictions = [x if random() > 0.15 else abs(1-x) for x in true_labels]
```

Write a function that takes the lists of predictions and labels and creates and returns a dictionary that contains the counts of true positives, true negatives, false positives, false negatives for the input lists. Your function should call the function you wrote for the last task.  

```
def test_func_list(predictions, labels):
  value_matrix = {}
  for pred,label in zip(predictions,true_labels):
    value = test_func(pred,label)
    if value in value_matrix:
      value_matrix[value] += 1
    else:
      value_matrix[value] = 1

  return value_matrix
```

Test your function using true_labels and predictions variables.

```
test_func_list(predictions,true_labels)
```

Write a function that takes the lists of predictions and labels and returns the accuracy of the predictions. Accuracy is defined as all correct predictions (TPs and TNs) divided by all predictions (TP+FP+TN+FN). Your function should call the function that counts the types of events (from the last exercise).

```
def accuracy(predictions, true_labels):
  result = test_func_list(predictions, true_labels)
  correct = result['tp']+result['tn']
  total = correct + result['fp'] + result['fn']
  return correct/total
```

Test your function using true_labels and predictions variables.

```
accuracy(predictions, true_labels)
```

Write a function that computes the MCC metric (refer for the formula: https://en.wikipedia.org/wiki/Matthews_correlation_coefficient).
You will need to use the `math` module to calculate the square root for the denominator.

Test your function using `true_labels` and `predictions` variables.

```
import math
def MCC_accuracy(predictions, true_labels):
  result = test_func_list(predictions, true_labels)
  numerator = result['tp']*result['tn']-result['fp']*result['fn']
  denominator = math.sqrt((result['tp']+result['fp'])*(result['tp']+result['fn'])*(result['tn']+result['fp'])*(result['tn']+result['fn']))
  return numerator/denominator
```
